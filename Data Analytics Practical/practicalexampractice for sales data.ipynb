```python
df=pd.read_csv("sales.csv")

#Create python numpy arrays
orderno = ((np.asarray(df['ORDERNUMBER'])).reshape(5,5))
sales = ((np.asarray(df['SALES'])).reshape(5,5))
#print(name)
#print(se)
# Create a pivot table
result = df.pivot(index='in',columns='col',values='SALES')
#print(result)
#Create an array to annotate the heatmap
labels = (np.asarray(["{0} \n {1:.0f}".format(symb,value)
                      for symb, value in zip(orderno.flatten(),
                                               sales.flatten())])
         ).reshape(5,5)
# Define the plot
fig, ax = plt.subplots(figsize=(5,5))

# Add title to the Heat map
title = "Sales"

# Set the font size and the distance of the title from the plot
plt.title(title,fontsize=18)
ttl = ax.title

# Remove the axes
ax.axis('off')

# Use the heatmap function from the seaborn package
sns.heatmap(result, annot=labels, fmt="",cmap='RdYlGn',linewidths=0.30,ax=ax)

# Display the Heatmap
plt.show()
```
